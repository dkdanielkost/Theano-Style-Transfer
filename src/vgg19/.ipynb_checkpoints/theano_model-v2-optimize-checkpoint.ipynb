```python
IMAGE_W = 224
MEAN_VALUES = np.array([104, 117, 123]).reshape((3,1,1))

def prep_image(im):
    if len(im.shape) == 2:
        im = im[:, :, np.newaxis]
        im = np.repeat(im, 3, axis=2)
    h, w, _ = im.shape
    if h < w:
        im = skimage.transform.resize(im, (IMAGE_W, w*IMAGE_W/h), preserve_range=True)
    else:
        im = skimage.transform.resize(im, (h*IMAGE_W/w, IMAGE_W), preserve_range=True)

    # Central crop
    h, w, _ = im.shape
    im = im[h//2-IMAGE_W//2:h//2+IMAGE_W//2, w//2-IMAGE_W//2:w//2+IMAGE_W//2]
    
    rawim = np.copy(im).astype('uint8')
    
    # Shuffle axes to c01
    im = np.swapaxes(np.swapaxes(im, 1, 2), 0, 1)
    
    # Convert RGB to BGR
    im = im[::-1, :, :]

    im = im - MEAN_VALUES
    return rawim, (im[np.newaxis]).astype('float32')

photo = imread('images/tubingen.jpg')
rawim, content = prep_image(photo)
```

```python
import skimage
import scipy
from skimage.io import imread
from skimage.transform import resize
mean_pixel = np.array([103.939, 116.779, 123.68])

def prepare_img(img):
    img = resize(img, (224, 224), preserve_range=True)
    original_img = np.copy(img).astype('uint8')
    img = img.transpose(2, 0 , 1)

    # Convert RGB to BGR
    img = img[::-1, :, :]

    #img = img - mean_pixel.reshape(3, 1, 1)

    return (img[np.newaxis]).astype(np.float32)

def recover_image(img):
    img = np.copy(img[0])
    img += mean_pixel.reshape(3, 1, 1)

    img = img[::-1]
    return img.transpose(1, 2, 0).astype('uint8')

def gram_matrix(x):
    x = x.flatten(ndim=3)
    g = T.tensordot(x, x, axes=([2], [2]))
    return g

def gram_scale(mat):
    N = mat.shape[1]
    M = mat.shape[2] * mat.shape[3]
    return 1.0/(4*N**2*M**2)

def total_variation_loss(x):
    return (((x[:,:,:-1,:-1] - x[:,:,1:,:-1])**2 + (x[:,:,:-1,:-1] - x[:,:,:-1,1:])**2)**1.25).sum()
```

```python
# wn = theano.shared(value=white_noise, name='wn', borrow=True)
vgg19 = VGG19(input_image_shape=content.shape)


get_content_layer = theano.function(inputs=[vgg19.input],
                                    outputs=vgg19.conv4_2.output)

get_style_layers = theano.function(inputs=[vgg19.input],
                                    outputs=[vgg19.conv1_1.output, vgg19.conv2_1.output,
                                             vgg19.conv3_1.output, vgg19.conv4_1.output,
                                             vgg19.conv5_1.output])
```

```python
#get_params = theano.function(inputs=[vgg19.input], outputs = [vgg19.conv1_1.params])

######### Stuff above this only needs to be run once ##########
#sum(np.mean(np.square(L1[i]+L2[i])) for i in range(len(L1)))
#might be just wn
sty_out = get_style_layers(style)
cont_lay = get_content_layer(content)

wn = theano.shared(value=white_noise, name='wn', borrow=False)
wn_cont = T.dtensor4('wn_cont')
cont = T.dtensor4('cont')
gram = T.dtensor4('gram')
sty1 = T.dtensor4('sty1')
sty2 = T.dtensor4('sty2')
sty3 = T.dtensor4('sty3')
sty4 = T.dtensor4('sty4')
sty5 = T.dtensor4('sty5')

cont = T.dtensor4('cont')

cont_loss = 0.001*0.5*T.sum(T.sqr(vgg19.conv4_2.output-cont))

style_loss = 0.2e6*(T.sum(T.sqr(gram_matrix(sty1)-gram_matrix(vgg19.conv1_1.output)))*gram_scale(gram_matrix(sty1)) +
    T.sum(T.sqr(gram_matrix(sty2)-gram_matrix(vgg19.conv2_1.output)))*gram_scale(gram_matrix(sty2)) +
    T.sum(T.sqr(gram_matrix(sty3)-gram_matrix(vgg19.conv3_1.output)))*gram_scale(gram_matrix(sty3)) +
    T.sum(T.sqr(gram_matrix(sty4)-gram_matrix(vgg19.conv4_1.output)))*gram_scale(gram_matrix(sty4)) +
    T.sum(T.sqr(gram_matrix(sty5)-gram_matrix(vgg19.conv5_1.output)))*gram_scale(gram_matrix(sty5)))

#cost = cont_loss
cost = 0.1e-7*total_variation_loss(vgg19.input) + cont_loss + style_loss

img_grad = T.grad(cost, vgg19.input)
```

```python
# Theano functions to evaluate loss and gradient
f_loss = theano.function([vgg19.input, cont, sty1, sty2, sty3, sty4, sty5], cost, on_unused_input='ignore')
f_grad = theano.function([vgg19.input, cont, sty1, sty2, sty3, sty4, sty5], img_grad, on_unused_input='ignore')

# Helper functions to interface with scipy.optimize
def eval_loss(x0):
    x0 = x0.reshape((1, 3, 224, 224)).astype(theano.config.floatX)
    wn.set_value(x0)
    return f_loss(wn.get_value(), cont_lay, sty_out[0], sty_out[1], sty_out[2], sty_out[3], sty_out[4]).astype('float64')

def eval_grad(x0):
    x0 = x0.reshape((1, 3, 224, 224)).astype(theano.config.floatX)
    wn.set_value(x0)
    return np.array(f_grad(wn.get_value(), cont_lay, sty_out[0], sty_out[1], sty_out[2], sty_out[3], sty_out[4])).flatten().astype('float64')
```

```python
wn.set_value(white_noise)

x0 = wn.get_value().astype('float64')
xs = []
losses = []
grads = []
xs.append(x0)
cont_lay = get_content_layer(content)
sty_out = get_style_layers(style)

# Optimize, saving the result periodically
for i in range(9):
    print(i)
    scipy.optimize.fmin_l_bfgs_b(eval_loss, x0.flatten(), fprime=eval_grad, maxfun=70)
    x0 = wn.get_value().astype('float64')
    xs.append(x0)
    losses.append(eval_loss)
```
